```python
import pandas as pd
```

```python

def create_df(path):
    """
    Reads the esoteric file format that stores eeg data and metadata
    ---

    """
    data = {}
    with open(path) as fp:

        line = fp.readline() # first line
        name = line.split('# ')[-1].split('.rd')[0] # subject id
        cnt = 1

        while line:

            line = fp.readline()
            cnt += 1

            if cnt == 4:
                trial_num = line.split('trial ')[-1].strip()

            elif cnt > 4:


                if line == '':
                    continue
                elif line.split(': ')[-1][0] == '#':
                    probe = line.split('# ')[-1].split(' ')[0]
                    data['trial']= trial_num
                    data['subject'] = name
                    data[probe] = []

                else:
                    data[probe].append(line.split(' ')[-1].strip())
                    
    return pd.DataFrame(data)
```

```python
import os
rootdir = './'

df = pd.DataFrame()
cnt = 1
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        if '.rd' in os.path.join(subdir, file):
            df = df.append(create_df(os.path.join(subdir, file)))
            print(f'Loaded the # {cnt} trial')
            cnt += 1
```

```python
df.to_csv('alcohol_eeg.csv')
```
