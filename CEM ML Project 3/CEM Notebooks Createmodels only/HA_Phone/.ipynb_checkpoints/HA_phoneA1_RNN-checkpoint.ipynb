```python
import pandas as pd
import numpy as np
```

```python
filen = "Phones_accelerometer.csv"
df1 = pd.read_csv(filen)
df1.head()
```

```python
filen1 = "Phones_gyroscope.csv"
df2 = pd.read_csv(filen1)
df2.head()
```

```python
df1.append(df2)
df1.head()
```

```python
df1.count()
```

```python
df1['gt'].unique()
```

```python
to_stay = ['stand','sit', 'walk', 'stairsup', 'stairsdown', 'bike']
df = df1[df1['gt'].isin(to_stay)]
df.dropna()
df.head()
```

```python
df.count()
```

```python
df['gt'].unique()
```

```python
y = df["gt"]
y
```

```python
X = df.iloc[:,3:6]
X.head()
```

```python
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.utils import to_categorical
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.utils import to_categorical

X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1, stratify=y)
X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)


# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)

# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
X_test_scaled
```

```python
X_test
```

```python
X_test["Status"] = y_test
X_test.head()
```

```python
X_test.to_csv("HA_test_phone.csv", encoding='utf-8')
```

```python
X_train["Status"] = y_train
X_train.head()
```

```python
X_train.to_csv("HA_train_phone.csv", encoding='utf-8')
```

```python
X_train_scaled
```

```python
parameters = len(X_train_scaled[0,:])
print (parameters)
```

```python
#X_train_scaled = X_train_scaled[:,1:parameters+1]
#X_train_scaled
```

```python
X_train_scaled = np.reshape(X_train_scaled, (X_train_scaled.shape[0],1,X_train_scaled.shape[1]))
X_train_scaled = np.array(X_train_scaled)
X_train_scaled
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

s = 12

# Create model and add layers
model = Sequential()
model.add(LSTM(s, input_shape=(None, parameters), return_sequences= False))
model.add(Dense(6, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
model.fit(X_train_scaled, y_train_categorical, epochs=6, batch_size=40)
```

```python
X_test_scaled = np.reshape(X_test_scaled, (X_test_scaled.shape[0],1,X_test_scaled.shape[1]))
X_test_scaled = np.array(X_test_scaled)
X_test_scaled
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes(X_test_scaled[:5])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:5])}")
```
