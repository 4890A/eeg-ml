```python
import gzip
import pandas as pd
```

```python
def readfile(fname1):
    
    data = {}
    
    with gzip.open(fname1,'r') as fin:
        
        i = 0
        firstChannel = 0
        
        for line in fin:
            y = line.split()
            
            if i == 0:
                subject = y[1].split()[0].decode('ascii').split(".")[0]
                data["Subject"] = []
                data["Trial"] = []
                data["Index"] = []
            elif i > 3:
                a1 = y[0].split()[0].decode('ascii')
                a2 = y[1].split()[0].decode('ascii')
                a3 = y[2].split()[0].decode('ascii')
                a4 = y[3].split()[0].decode('ascii')
                if a3 == "chan":
                    channel = a2
                    data[channel] = []
                    firstChannel += 1
                elif a2 == channel:
                    if firstChannel == 1:
                        data["Subject"].append(subject)
                        data["Trial"].append(a1)
                        data["Index"].append(i-4)
                        data[channel].append(a4)
                    else:
                        data[channel].append(a4)
            i += 1
    return pd.DataFrame(data)
                 
```

```python
fname1 = "co2a0000364.rd.030.gz"

df = readfile(fname1)
df.head()
```

```python
import os
rootdir = './complete data/SMNI_CMI_TRAIN'

df = pd.DataFrame()
cnt = 1
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        if '.rd' in os.path.join(subdir, file):
            df = df.append(readfile(os.path.join(subdir, file)))
            print(f'Loaded the # {cnt} trial')
            cnt += 1
```

```python
df.to_csv("CEM_train_Alcoh.csv", encoding='utf-8')
```
