```python
#connect to the database
db = sqlite3.connect(database)

query = "SELECT * FROM Alc_test1;"
df = pd.read_sql_query(query,db)
db.close()
df.head()
```

```python
y1 = df["Status"]
y2 = []
for s in y1:
    if 'alcoholic' in s:
        y2.append(1)
    else:
        y2.append(0)

y4 = pd.DataFrame({"label":y2})
y_test = y4["label"]
```

```python
X_test = df.iloc[:,1:65]
X_test.head()
```

```python
#Load the Scaled function for LR model
from sklearn.externals.joblib import dump, load
X_scaler = load('Alc_Brain_LR_scaler.bin')
```

```python
#Scales X_test data
X_test_scaled = X_scaler.transform(X_test)  
```

```python
#Load the Lineal regression model, and print the accuracy score
import pickle
filename = "Alc_Brain_LR.sav"
model = pickle.load(open(filename, 'rb'))
result = model.score(X_test_scaled, y_test)
print(result)
```
