```python
def classification(run, event):
    
    if run == '01':
        classification = 'BL Eyes Open'
        
    elif run == '02':
        classification = 'BL Eyes Closed'
        
    elif run == '03':
        if event == 'T0':
            classification = 'Rest'
        elif event == 'T1':
            classification = 'Motion Left Fist'
        else: 
            classification = 'Motion Right Fist'
            
    elif run == '04':
        if event == 'T0':
            classification = 'Rest'
        elif event == 'T1':
            classification = 'Imagine Left Fist'
        else:
            classification = 'Imagine Right Fist'
    
    elif run == '05':
        if event == 'T0':
            classification = 'Rest'
        elif event == 'T1':
            classification = 'Motion Both Fist'
        else:
            classification = 'Motion Both Feet'
    
    else:
        if event == 'T0':
            classification = 'Rest'
        elif event == 'T1':
            classification = 'Imagine Both Fist'
        else:
            classification = 'Imagine Both Feet'
            
    return classification
```

```python
classification('03', 'T2')
```

```python
classification_list = []
for index, row in df_filter.head().iterrows():
    run = row['name'][-2:]
    event = row['event']
    classification_list.append(classification(run, event))
```
